```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import matplotlib.dates as mdates
```

```python
# Import data (Make sure to parse dates. Consider setting index column to 'date'.)
df = pd.read_csv("fcc-forum-pageviews.csv",parse_dates=["date"])
df
```

```python
# Setting index column to "date"
df.index = df["date"]
df.drop(columns="date",inplace=True)
df
```

```python
# Clean data (Remove outliers)
df = df[(df["value"] >= df["value"].quantile(0.025)) & (df["value"] <= df["value"].quantile(0.975))]
df
```

```python
# Draw line plot and save it
fig, ax = plt.subplots(figsize=(12,5))
plt.plot(df.index, df["value"],"r",linewidth=1)
#showing only year-month and every 6th month's dates on x label
date_form = mdates.DateFormatter("%Y-%m")
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.MonthLocator((1,7)))
plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
plt.xlabel("Date")
plt.ylabel("Page Views")
fig.savefig('line_plot.png')
plt.show()
```

```python
# Copy and modify data for monthly bar plot
df_bar = df.copy()
df_bar['month'] = pd.DatetimeIndex(df_bar.index).month
df_bar['year'] = pd.DatetimeIndex(df_bar.index).year
df_bar=df_bar.groupby(["year","month"])["value"].mean().unstack()
df_bar
```

```python
# Draw bar plot
fig = df_bar.plot.bar(legend=True)
fig.legend(['January', 'February', 'March', 
            'April', 'May', 'June', 'July', 
            'August', 'September', 'October', 
            'November', 'December'],prop={'size': 7}, title='Months')
plt.xlabel("Years")
plt.ylabel("Average Page Views")
plt.tight_layout()
fig=fig.figure
fig.savefig('bar_plot.png')
```

```python
# Prepare data for box plots (this part is done!)
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['year'] = [d.year for d in df_box.date]
df_box['month'] = [d.strftime('%b') for d in df_box.date]
df_box
```

```python
#in order to start the months with Jan, we need to sort the months.
#firstly extract the months out of the dates and sort them
df_box["num_of_month"] = df_box["date"].dt.month 
df_box["num_of_month"]
df_box = df_box.sort_values("num_of_month")
df_box
```

```python
# Draw box plots (using Seaborn)
fig, axes = plt.subplots(ncols=2,figsize=(12,6))
axes[0]=sns.boxplot( ax=axes[0],data=df_box, x="year", y="value")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Page Views")
axes[0].set_title("Year-wise Box Plot (Trend)")

axes[1]=sns.boxplot( ax=axes[1], data=df_box, x="month", y="value")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Page Views")
axes[1].set_title("Month-wise Box Plot (Seasonality)")
```
